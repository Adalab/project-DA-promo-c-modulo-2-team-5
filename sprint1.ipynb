Creamos una funci√≥n para extraer toda la informaci√≥n de nuetros archivos de forma r√°pida:

```python
def conocer_archivo(df):
    print(f"La forma de mi archivo es: {df.shape}")
    print("---------------------------------------")
    print(f"Los principales datos estadisticos de mi archivo son: {df.describe()}")
    print("---------------------------------------")
    print(f"La informacion de mi archivo es: {df.info()}")
    print("---------------------------------------")
    print(f"Las columnas de mi archivo son: {df.columns}")
    print("---------------------------------------")
    print(f"Los tipos de datos de mi archivo son: {df.dtypes}")
    print("---------------------------------------")
    print(f"Tenemos estos valores nulos : {df.isnull().sum()}")
    print("---------------------------------------") 
```

```python
conocer_archivo(txt_limpia)
```
