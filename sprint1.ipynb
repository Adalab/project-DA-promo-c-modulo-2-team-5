Creamos una función para extraer toda la información de nuetros archivos de forma rápida:

```python
def conocer_archivo(df):
    print(f"La forma de mi archivo es: {df.shape}")
    print("---------------------------------------")
    print(f"Los principales datos estadisticos de mi archivo son: {df.describe()}")
    print("---------------------------------------")
    print(f"La informacion de mi archivo es: {df.info()}")
    print("---------------------------------------")
    print(f"Las columnas de mi archivo son: {df.columns}")
    print("---------------------------------------")
    print(f"Los tipos de datos de mi archivo son: {df.dtypes}")
    print("---------------------------------------")
    print(f"Tenemos estos valores nulos : {df.isnull().sum()}")
    print("---------------------------------------") 
```

```python
conocer_archivo(txt_limpia)
```

```python
# Transformamos la columna index del archivo txt, que era de tipo object, a integer, para poder hacer el merge:

txt_limpia["index"] = txt_limpia["index"].astype(int)
```

```python
# Hacemos el merge:

txt_xml_merge = data_xml_clean_xml.merge(txt_limpia, how='inner', on="index")
```

```python
# Llevamos el dataframe que vamos a usar para la estadística a un archivo csv:

txt_xml_merge.to_csv('txt_xml_merge.csv')
```
